```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import os
from PIL import Image

names = ["michelangelo buonarotti", "vincent van gogh", "edvard munch"]
pieces = {"michelangelo buonarotti": ["die heilige familie", "die erschaffung adams"], 
          "vincent van gogh": ["selbstbildnis","das nachtcafe", "gemüsegärten auf dem montmartre", "die kartoffelesser", "kornfeld mit krähen", "das weiße haus bei nacht", "porträt des dr. gachet"], 
          "edvard munch": ["der schrei", "der kuss", "das kind und der tod", "selbstbildnis mit spanischer grippe", "zwei menschen. die einsamen", "abend auf der karl johans gate"]}
for dkey, dvalues in pieces.items():
    for lvalue in dvalues:
        element = dkey +' '+ lvalue
        print(element)
        #opens chrome and confirms conditions
        driver = webdriver.Chrome('C:/Users/Huckleberry Finn/Unternehmenssoftware/chromedriver.exe')
        driver.get('https://www.google.de/imghp')
        button = driver.find_element_by_css_selector('#L2AGLb > div')
        button.click()
        #types subject in search field
        box = driver.find_element_by_xpath('/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div[2]/div[2]/input')
        #element = 'edvard munch der schrei'
        box.send_keys(element)
        box.send_keys(Keys.ENTER)
        #open style "original"
        button = driver.find_element_by_css_selector('#i6 > div.FAZ4xe.FoDaAb > span > span:nth-child(1) > div > a')
        button.click()
        #Will keep scrolling down the webpage until it cannot scroll no more
        last_height = driver.execute_script('return document.body.scrollHeight')
        while True:
            driver.execute_script('window.scrollTo(0,document.body.scrollHeight)')
            new_height = driver.execute_script('return document.body.scrollHeight')
            try:
                driver.find_element_by_xpath('//*[@id="islmp"]/div/div/div/div/div[5]/input').click()
            except:
                pass
            if new_height == last_height:
                break
            last_height = new_height
        #checks if folder exists, if not create it
        os.chdir('E:/08_Projekte/Unternehmenssoftware/pictures/')
        if not os.path.exists(dkey):
            os.makedirs(dkey)
            
        os.chdir('E:/08_Projekte/Unternehmenssoftware/pictures/'+dkey+'/')
        if not os.path.exists(lvalue):
            os.makedirs(lvalue)
        
        os.chdir('E:/08_Projekte/Unternehmenssoftware/pictures/'+dkey+'/'+lvalue+'/')
        if not os.path.exists("test"):
            os.makedirs("test")
        
        os.chdir('E:/08_Projekte/Unternehmenssoftware/pictures/'+dkey+'/'+lvalue+'/')
        if not os.path.exists('train'):
            os.makedirs('train')
                      
        #takes screenshots from every picture in the range    
        for i in range(1,15):
            try:
                driver.find_element_by_tag_name('body').send_keys(Keys.CONTROL + Keys.HOME)
                time.sleep(0.2)
                driver.find_element_by_xpath('//*[@id="islrg"]/div[1]/div['+str(i)+']/a[1]/div[1]/img').screenshot('E:/08_Projekte/Unternehmenssoftware/pictures/'+dkey+'/'+lvalue+'/('+str(i)+').png')
            except:
                pass
           
        os.chdir('E:/08_Projekte/Unternehmenssoftware/pictures/'+dkey+'/'+lvalue+'')
        #path = os.getcwd()
        #print(path)
        driver.close()
        #resizes all pictures
        for j in range(1,15):
            PATH = '('+str(j)+').png'
            if os.path.isfile(PATH) and os.access(PATH, os.R_OK):
                if j <= 10:
                    testtrain = 'train'
                else: 
                    testtrain = 'test'
                image = Image.open('('+str(j)+').png')
                new_image = image.resize((224, 224))
                rgb_new_image = new_image.convert('RGB')
                rgb_new_image.save(''+testtrain+'/('+str(j)+').jpg')
                os.remove('('+str(j)+').png')
            else:
                print("Either the file is missing or not readable")
```
